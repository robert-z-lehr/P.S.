### Summary:
- Statistics helps us make intelligent judgments and informed decisions in the presence of uncertainty and variation. Without uncertainty or variation, there would be little need for statistical methods or statisticians.
- Statistics includes methods to ensure studied variations are not due to eternal confounding factors, as a safeguard to prevent incorrect conclusions from being made.
    - `data:` Collections of facts; results from making observations on a single variable or simultaneously on 2+ variables.
- Statistics provides methdos for describing and summarizing data and for drawing conclusions based on information contained in the data.
    - `population (of interest):` A well-defined collection of objects constitutes a population of interest. Usually, only a select few characteristics of the population are considered to be valuable for the respective study.
    - `census:` When all desired information is available for all objects in the population, this is considered to be a census. Scarcity of time, money, or other resources usually make obtaining a census impractical or infeasible.
    - `sample:` A subset of the population, selected in some prescribed manner.
    - `characteristic (categorical):` Non-numerical characteristics, such as gender or type of malfunction
    - `characteristic (numerical):` Numerical characteristics, such as age or diameter.
    - `variable:` Any characteristic whose value may change from one object to another in the population.
    - `univariate dataset:` Consists of observations on a single variable. (e.g. - determine the type of transmission on a sinel vehicle. Automative or Manual)
    - `bivariate dataset:` Consists of observations made on each of two variables. (e.g. - measuring height and weight)
    - `multivariate dataset:` Consists of observations made on more than one variable (bivariate datasets being a type o multivariate datasets). It is common for some variables to be numerical and others to be categorical.
- Branches of statistics: Descriptive statistics and inferential statistics.
    - `descriptive statistics:` The important features of the collected data are identified, summarized, and described. Descriptive statistical methods are often graphical in nature (e.g. - constructing histograms, boxplots, and scatterplots). Other descriptive methods involved calculating summary measures (i.e. - mean, median, variance, standard deviation, and correlation coefficients). The importance of these methods are when there is a need to discover the prominent features of a population, which are otherwise difficult to determine when dealing with raw, unordered data. Determining what value is typical (a representative value), where values are concentrated around or dispersed, or whether there are gaps in the data, are difficult to discern without descriptive statistical methods. Descriptive statistics are used in the development of inference.
    - `inferential statistics:` Once a sample of a population is collected, conclusions for the entire population are inferred, making the sample a means to an end, rather than the end itself. Techniques for generalizing from a sample to a population are considered inferential statistical techniques.
    - `confidence interval / interval estimate:` Summary measures  (e.g. - mean, median, variance, standard deviation, etc.) can be be estimated within a confidence interval /  interval estimate.
    - `lower prediction bound:` The lower bound for the likelihood of a measurement occurring. (e.g. - It can be measured witha  high degree of confidence that the anyone vehicle will be driving faster than 50 mph on a highway)
- Important inferential procedures include: point estimation, hypothesis testing, and estimation by confidence intervals.
    - `Probability:` Understanding probability leads to a better understanding of how inferential procedures are developed and used. In a probability problem, properties of the population under study are assumed known and questions regarding a sample taken from the population are posed and answered.
    - `Statistics`: Characteristics of a sample are known, and this information enables the ability to draw conclusions about the population.
    - `Probability vs. Inferential Statistics:` Probabilty reasons from the population to the sample (deductive reasoning), whereas inferential statistics reasons from the sample to the population (inductive reasoning).
    - `Probability Example:` 50% of all drivers will use their seatbelt. How likely is it that out of a sample of 100 drivers, at least 70 will use their seatbelt?
    - `Inferential Statistics Examples`: A sample of 100 drivers show that 65 use their seatbelts. Does this provide substantial evidence for concludin that more than 50% of all such drivers in this area (the population) use their seatbelts?
    - [__???__] `Hypothetical / Conceptual Population`: When the likely values of the characteristics of the full population cannot be readily determined, a conceptualized population can inferred from all possible values in chosen environmental conditions.
    - `Enumerative Studies:` Interest is focused on a finite, identifiable, unchanging collection of individuals or objects that make up a population.
    - `Analytic Studies:` Studies that are not enumerative in nature. Such studies are often carried out with the objective of improving a future
product by taking action on a process of some sort (e.g., recalibrating equipment or adjusting the level of some input such as the amount of a catalyst).
    - `Sampling frame:` a listing of the individuals or objects to be sampled
    - `simple random sample:` When data collection entails selecting individuals or objects from a frame, the simplest method for ensuring a representative selection is to take a simple random sample. This is one for which any particular subset of the specified size (e.g., a sample of size 100) has the same chance of being selected.
    - `stratfied sampling:` Sometimes alternative sampling methods can be used to make the selection process easier, to obtain extra information, or to increase the degree of confidence in conclusions. One such method, stratified sampling, entails separating the population units into nonoverlapping groups and taking a sample from each one. For example, a manufacturer of DVD players might want information about customer satisfaction for units produced during the previous year. If three different models were manufactured and sold, a separate sample could be selected from each of the three corresponding strata. This would result in information on all three models and ensure that no one model was over- or underrepresented in the entire sample.
    - `a "convenience" sample:` Frequently a “convenience” sample is obtained by selecting individuals or objects without systematic randomization. As an example, a collection of bricks may be stacked in such a way that it is extremely difficult for those in the center to be selected. If the bricks on the top and sides of the stack were somehow different from the others, resulting sample data would not be representative of the population. Often an investigator will assume that such a convenience sample approximates a random sample, in which case a statistician’s repertoire of inferential methods can be used; however, this is a judgment call.
    - `mutually excluaive = pairwise disjoint`

### Question 2:
__How much more__ __*money*__ will I make by getting a __PhD__ rather than an __MS__?

- STEP 1 (Identify the problem): To calculate event 1 (how much more money I will make) based on event 2 (getting a PhD degree) occurring rather than based on event 3 (getting an MS degree) occurring.

- STEP 2 (Decompose the problem):
    - STEP 2.1: PD(E1) = The probability distribution of event 1 (how much money I will make based on getting a PhD degree)
    - STEP 2.2: PD(E2) = The probability distribution of event 2 (how much more money I will make based on getting a PhD degree)
    - STEP 2.3: PD(E1 - E2) = The probability distribution of how much more money I will make based on getting a PhD degree rather than an MS degree.

- STEP 3 (Structure the problem): 
    - STEP 3.1: Determine how to calculate the PD(E1).
    - STEP 3.2: Calculate the PD(E1).
    - STEP 3.3: Visualize the PD(E1) (if time allows).
    - STEP 3.4: Determine how to calculate the PD(E2).
    - STEP 3.5: Calculate the PD(E2).
    - STEP 3.6: Visualize the PD(E2) (if time allows).
    - STEP 3.7: Determine how to calculate the PD(E1 - E2).
    - STEP 3.8: Calculate the PD(E1 - E2).
    - STEP 3.9: Visualize the PD(E1 - E2) (if time allows).
    - STEP 3.10 Calculate the difference in agerage and standard deviations between the two distributions to determine how much more money will be made on average.

- STEP 4 (Solve the problem): *Note: the fillowing assumptions are used for educational purposes*

````python
# STEP 4:
# Create a random number generator
# Create a random number generator
rng = np.random.default_rng()

# Generate normal distributions
probability_distribution_money_from_PhD = rng.normal(60000, 10000, 100)
probability_distribution_money_from_Masters = rng.normal(60000, 10000, 100)

# Calculate means
mean_PhD = np.mean(probability_distribution_money_from_PhD)
mean_Masters = np.mean(probability_distribution_money_from_Masters)

# Calculate the difference between the distributions
difference = probability_distribution_money_from_PhD - probability_distribution_money_from_Masters

# Calculate the mean and standard deviation of the differences
mean_difference = np.mean(difference)
std_difference = np.std(difference)

print(f"Mean annual income for PhD: ${mean_PhD:.2f}")
print(f"Mean annual income for Masters: ${mean_Masters:.2f}")
print(f"Mean difference in annual income: ${mean_difference:.2f}")
print(f"Standard deviation of the difference: ${std_difference:.2f}")

# Plot the distributions
plt.hist(probability_distribution_money_from_PhD, bins=30, alpha=0.5, label='PhD')
plt.hist(probability_distribution_money_from_Masters, bins=30, alpha=0.5, label='Masters')

# Add labels and title
plt.xlabel('Annual Income')
plt.ylabel('Frequency')
plt.title('Probability Distribution of Annual Income')
plt.legend()

# Show the plot
plt.show()

# Plot the difference distribution
plt.hist(difference, bins=30, alpha=0.5, label='Difference (PhD - Masters)')

# Add labels and title
plt.xlabel('Difference in Annual Income')
plt.ylabel('Frequency')
plt.title('Distribution of Income Difference')
plt.legend()

# Show the plot
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind

# Create a random number generator
rng = np.random.default_rng()

# Generate normal distributions
probability_distribution_money_from_PhD = rng.normal(60000, 10000, 100)
probability_distribution_money_from_Masters = rng.normal(60000, 10000, 100)

# Calculate means
mean_PhD = np.mean(probability_distribution_money_from_PhD)
mean_Masters = np.mean(probability_distribution_money_from_Masters)

# Perform the two-sample t-test
t_stat, p_value = ttest_ind(probability_distribution_money_from_PhD, probability_distribution_money_from_Masters)

print(f"Mean annual income for PhD: ${mean_PhD:.2f}")
print(f"Mean annual income for Masters: ${mean_Masters:.2f}")
print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.5f}")

# Compare the p-value to the significance level
if p_value < 0.05:
    print("There is a statistically significant difference between the incomes from a PhD and a Masters degree.")
else:
    print("There is no statistically significant difference between the incomes from a PhD and a Masters degree.")

# Plot the distributions
plt.hist(probability_distribution_money_from_PhD, bins=30, alpha=0.5, label='PhD')
plt.hist(probability_distribution_money_from_Masters, bins=30, alpha=0.5, label='Masters')

# Add labels and title
plt.xlabel('Annual Income')
plt.ylabel('Frequency')
plt.title('Probability Distribution of Annual Income')
plt.legend()

# Show the plot
plt.show()

# Plot the difference distribution
plt.hist(probability_distribution_money_from_PhD - probability_distribution_money_from_Masters, bins=30, alpha=0.5, label='Difference (PhD - Masters)')

# Add labels and title
plt.xlabel('Difference in Annual Income')
plt.ylabel('Frequency')
plt.title('Distribution of Income Difference')
plt.legend()

# Show the plot
plt.show()
````

## Detailed Pros, Cons, and Use Cases

### Direct Instruction
- **Pros:** Highly structured, clear objectives, efficient for teaching specific facts and procedures.
- **Cons:** Can be passive for students, may not develop critical thinking skills.
- **Best Use Cases:** Teaching foundational knowledge, introducing new concepts.

### Problem-Based Learning (PBL)
- **Pros:** Develops problem-solving and critical thinking skills, engages students with real-world problems.
- **Cons:** Time-consuming, may be challenging for students with less prior knowledge.
- **Best Use Cases:** Complex subjects, real-world applications, Engineering, Medicine, Business.

### Project-Based Learning (PjBL)
- **Pros:** Encourages collaboration, applies knowledge to practical projects.
- **Cons:** Requires significant planning, assessment can be challenging.
- **Best Use Cases:** Subjects needing practical application, STEM fields, Art and Design.

### Inquiry-Based Learning
- **Pros:** Encourages curiosity and independent thinking, develops research skills.
- **Cons:** Can be difficult to manage, requires skilled facilitation.
- **Best Use Cases:** Science and Research, developing investigative skills.

### Cooperative Learning
- **Pros:** Promotes social interaction and teamwork, encourages diverse perspectives.
- **Cons:** Group dynamics can be challenging, may not engage all students equally.
- **Best Use Cases:** Group projects, collaborative research, peer learning.

### Flipped Classroom
- **Pros:** Maximizes classroom interaction time, allows students to learn at their own pace.
- **Cons:** Requires access to technology, preparation can be time-intensive.
- **Best Use Cases:** Subjects requiring discussion and application, Higher education.

### Blended Learning
- **Pros:** Combines the best of online and face-to-face learning, flexible and scalable.
- **Cons:** Requires significant resources, may be challenging to integrate seamlessly.
- **Best Use Cases:** Higher education, professional development.

### Experiential Learning
- **Pros:** Hands-on, practical experience, enhances retention through active participation.
- **Cons:** Resource-intensive, may not cover all theoretical aspects.
- **Best Use Cases:** Vocational training, internships and practicums, laboratory-based courses.

### Differentiated Instruction
- **Pros:** Addresses diverse learning needs, can improve student engagement.
- **Cons:** Requires extensive preparation, can be challenging to implement in large classes.
- **Best Use Cases:** Inclusive education, mixed-ability classrooms.

### Constructivist Learning
- **Pros:** Encourages active learning and critical thinking, builds on students' prior knowledge.
- **Cons:** Can be less structured, may be challenging for some students.
- **Best Use Cases:** Subjects where understanding context is crucial, Humanities and Social Sciences.

### Socratic Method
- **Pros:** Develops critical thinking and dialogue skills, engages students deeply in content.
- **Cons:** Can be intimidating for some students, requires skilled facilitation.
- **Best Use Cases:** Philosophy and Ethics, Law and Political Science, Literature and History.

### Gamification
- **Pros:** Increases motivation and engagement, makes learning fun and interactive.
- **Cons:** Can be distracting, not suitable for all types of content.
- **Best Use Cases:** Elementary education, training programs, language learning.

### Scaffolding
- **Pros:** Provides structure and support, helps build student confidence.
- **Cons:** Can be time-consuming, may lead to dependence on the teacher.
- **Best Use Cases:** Subjects requiring step-by-step learning, Early childhood education, Special education.

### Mastery Learning
- **Pros:** Ensures students achieve a high level of understanding, focuses on individual progress.
- **Cons:** Can be time-consuming, may require additional resources.
- **Best Use Cases:** Subjects with clear, sequential learning goals, Math and Science.
